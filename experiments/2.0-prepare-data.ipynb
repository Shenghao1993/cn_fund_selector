# I. Prepare Fund Profile Data

```python
etf_df = pd.read_csv('etfdb/etfs_details_type_fund_flow.csv')
print(etf_df.shape)

etf_df.head().T
```

```python
etf_df = etf_df[['Symbol', 'ETF Name', 'Asset Class']]
```

```python
sector_dict = {}
for file in glob('etfdb/etfs_details_*.csv'):
    if file != 'etfdb/etfs_details_type_fund_flow.csv':
        sector = '_'.join(file.split('_')[2:-1])
        sector_etfs = pd.read_csv(file)
        for symbol in sector_etfs['Symbol'].values:
            if symbol not in sector_dict:
                sector_dict[symbol] = sector
            else:
                print(f"Multiple sectors found for fund: {symbol}")
```

ETFs of the three benchmark indices:
<ul>
    <li>S&P 500: SPY</li>
    <li>Dow Jones Industrial Average: DIA</li>
    <li>Nasdaq 100: QQQ</li>
</ul>

```python
SPY_sector = sector_dict.get('SPY')
DIA_sector = sector_dict.get('DIA')
QQQ_sector = sector_dict.get('QQQ')

print(f"SPY: {SPY_sector}")
print(f"DIA: {DIA_sector}")
print(f"QQQ: {QQQ_sector}")
```

```python
sector_dict['SPY'] = 'benchmark'
sector_dict['DIA'] = 'benchmark'
```

```python
etf_df['Sector'] = etf_df['Symbol'].apply(lambda symbol: sector_dict.get(symbol))
etf_df.head()
```
